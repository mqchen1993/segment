```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

from torch.utils.data import DataLoader
from segment.data import IRCAD2D
```

```python
PATH = '/Users/youngtodd/data/3DIRCADb1'
data = IRCAD2D(PATH, 'bone')
trainloader = DataLoader(data)
```

```python
w=25
h=25
fig=plt.figure()
#matplotlib.rcParams['figure.figsize'] = (30.0, 30.0)
columns = 2
rows = 6
i = 1
for idx, (data, mask) in enumerate(trainloader):
    data = data.squeeze(0)
    fig.add_subplot(rows, columns, i)
    plt.imshow(data)
    mask = mask.squeeze(0)
    fig.add_subplot(rows, columns, i+1)
    plt.imshow(mask)
    i += 2
    if idx == 5:
        break
    
plt.show()
```
